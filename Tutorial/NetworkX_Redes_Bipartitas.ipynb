```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_excel("Red Bipartita.xlsx")
print(df.head())
```

```python
[(x['Pais'],x['Categoria']) for indice,x in df[0:5].iterrows()]
```

```python
B=nx.Graph()
```

```python
B.add_nodes_from(df['Pais'], bipartite=0)
B.add_nodes_from(df['Categoria'], bipartite=1)
```

```python
B.number_of_nodes()
```

```python
B.add_weighted_edges_from(
    [(x['Pais'],x['Categoria'],x['Ventas']) for indice,x in df.iterrows()]
)
```

```python
B.number_of_edges()
```

```python
nx.degree(B)
```

```python
plt.figure(figsize=(10,10),dpi=70)
pos=nx.bipartite_layout(B, df['Pais'],align='horizontal')
nx.draw(B,pos,with_labels=True)
plt.show()
```

```python
nx.average_degree_connectivity(B)
```

```python
nx.density(B)
```

```python
B.number_of_edges()*2/(B.number_of_nodes()*(B.number_of_nodes()-1))
```

```python
nx.diameter(B)
```
